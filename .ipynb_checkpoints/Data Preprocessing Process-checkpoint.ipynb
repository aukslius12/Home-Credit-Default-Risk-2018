```python
# Libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('data/application_train.csv', index_col=0)
#train = train.sample(100000) #for faster calculations.
```

```python
## Identifying categorical variables.
# Removing obvious float64 variables.
train_tmp = train.select_dtypes(exclude='float64')

unique_lenghts = [len(train_tmp[col].unique()) for col in train_tmp.columns]
unique_lenghts = pd.DataFrame({
    'col_name': train_tmp.columns,
    'unique_len': unique_lenghts
})

# Inspect variables with large amount of unique values. 
col_names = unique_lenghts[unique_lenghts['unique_len'] > 20]['col_name']
#train_tmp[col_names] 

# DAYS_ variables and, of course, children, aren't categorical out of all that we have.
non_categorical_ints = ['DAYS_BIRTH', 'DAYS_EMPLOYED', 'DAYS_ID_PUBLISH', 'CNT_CHILDREN']
```

```python
## Replacing categorical NA's.
# Creating categorical part of train data.
train_categ = train_tmp.drop(non_categorical_ints, axis=1)

# Inspecting column's with NA's.
categ_na_cols = [] 
for col in train_categ.columns:
    if any(train_categ[col].isna()):
        print(col, '\n', train_categ[col].unique())
        categ_na_cols.append(col)

# Manually replacing missing categorical variables.
train_categ[categ_na_cols] = train_categ[categ_na_cols].fillna('not specified')

print(any(train_categ.apply(pd.Series.isna).apply(any))) # No NA's left.
```

```python
# XNA's? Will be useful when we get to continuous variables.
def find_xna(series):
    return any(series == 'XNA')
    
any(train_categ.apply(find_xna)) # Oh fk. UGHHHHHHHHHHH

# Replacing XNA's.
train_categ['ORGANIZATION_TYPE'] = train_categ['ORGANIZATION_TYPE'].replace('XNA', 'not specified')
```

```python
## Transforming categorical variables.
from sklearn.preprocessing import LabelEncoder

def transform_encode(series):
    series_transf = LabelEncoder().fit_transform(series)
    if len(series.unique()) > 1:
        dummies = pd.get_dummies(series)
        dummies.columns = [str(val) for val in dummies.columns.values] # For numericassfdasfdasdsa.
        dummies.columns = series.name + '_' + dummies.columns
        series_transf = dummies.iloc[:, 1:] # Avoiding the dummy variable trap.
    else:
        # 1 indicates the 2nd unique value, as 0 is always chosen as the first value.
        series_transf = pd.Series(series_transf, name=series.name + '_' + series.unique()[1])
    return series_transf

# Merging everything into a single dataframe.
train_categ_final = pd.DataFrame()
for col in train_categ.columns:
    train_categ_final = pd.concat([train_categ_final, transform_encode(train_categ[col])], axis=1)
```

```python
## Inspecting continuous NA's.
# Transforming HOURS_ into sin((2*pi*HOURS_/24)) to imitate daily application fluctuations.
# Another consideration would be to use it as categorical variables.
train['HOUR_APPR_PROCESS_START'] = np.sin((2*np.pi*train['HOUR_APPR_PROCESS_START'].values)/24)

train_cont = pd.concat([train.select_dtypes('float64'), train[non_categorical_ints]], axis=1, sort=False)
```

```python
# Inspecting the amount of NA's.
import missingno as msno
missingdata_df = train_cont.columns[train_cont.isnull().any()].tolist()
msno.bar(train_cont[missingdata_df], color="blue", log=True, figsize=(30,18))
```

```python
missing_name, missing = [], []
for col in train_cont.columns:
    missing_name.append(col)
    missing.append(sum(train_cont[col].isna())) # Thats a lot, actually.

missing_data = pd.DataFrame({
    'name': missing_name,
    'ammount': missing
})

pd.options.display.max_rows = 100
missing_data.sort_values(by='ammount', ascending=False)
# Judging by names, most of missing values in these variables have meaning. No common area? NA - house is
# cheap, no resale value. No days employed? NA - unemployed. etc. etc.
```

```python
def create_ismissing_df(df):
    '''
    Creates an additional dataframe which indicates which rows in previous data frame had missing data
    in their variables.
    '''
    df_new = pd.DataFrame()
    for col in df.columns:
        missing = df[col].isnull()
        df_new = pd.concat([df_new, pd.DataFrame({col + '_IS_NA': missing})], axis=1)
    return(df_new)


import fancyimpute  
def parametric_input(df):
    '''
    Using mice to fill missing continuous variables, adding a column to indicate that they were missing.
    '''
    # Filling values using mice.
    mice_matrix = fancyimpute.MICE().complete(df.values)
    mice_df = pd.DataFrame(mice_matrix)
    mice_df.columns = df.columns
    mice_df.index = df.index
    
    # Adding an indicator dataframe.
    ismissing_df = create_ismissing_df(df)
    return(pd.concat([mice_df, ismissing_df], axis=1))
```

```python
# Filling the values and finalizing the df
train_cont = parametric_input(train_cont)

print(any(train_cont.apply(pd.Series.isna).apply(any))) # No NA's left.
```
